```python
class Student:
  def __init__(self,name):
    self.name = name
  def study(self,hard=False):
    if hard:
      print("%s 학생은 열심히 공부합니다." %self.name)
    else :
      print("%s 학생은 공부합니다." %self.name)

s = Student('Heekyung')
s.study()
s.study(hard=True)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.arange(0,10,0.1)
y=np.sin(x)

plt.plot(x,y)
```

```python
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x,y_sin)
plt.plot(x,y_cos)
plt.xlabel('x sxis label')
plt.ylabel('y sxis label')
plt.title('sin and cos')
plt.legend(['sin' , 'cos'])

plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.title('sin')

plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('cos')

plt.show()
```
