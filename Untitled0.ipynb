```python
#결과
from collections import deque

graph = {
    '현빈': ['친구1', '친구2'],
    '친구1': ['친구3'],
    '친구2': ['친구4'],
    '친구3': [],
    '친구4': ['친구1', '친구5'],
    '친구5': ['친구7'],
    '친구7': []
}

def infection_bfs(graph, infected, max_depth=2):  # 2단계까지만 탐색
    queue = deque([(infected, 0)])
    visited = {infected}
    result = []

    while queue:
        person, level = queue.popleft()
        if level == max_depth:
            continue
        for neighbor in graph[person]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append((neighbor, level + 1))
                result.append((neighbor, level + 1))
    return result

infected_person = '현빈'
infected_list = infection_bfs(graph, infected_person)

print(f"감염자: {infected_person}\\n2단계 이내 감염 경로 추적 결과:")
for name, dist in infected_list:
    print(f" - {name} (거리 {dist})")
```
