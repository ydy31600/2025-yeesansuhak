```python
from collections import deque

def bfs_with_path(graph, start, max_depth):
    visited = {start}              # 1) 방문한 노드를 기록하는 집합
    dist = {start: 0}              # 2) 시작점으로부터의 거리(깊이)를 기록하는 딕셔너리
    parent = {start: None}         # 3) 경로 복원용: 이 노드를 발견하게 한 부모 노드
    q = deque([start])             # 4) BFS용 큐. 처음에는 시작 노드만 넣어둔다
    results = []                   # 5) 최종으로 리턴할 (노드, 깊이, 경로) 리스트

    while q:                       # 큐가 빌 때까지 반복 (BFS 기본 패턴)
        cur = q.popleft()          # 큐에서 현재 노드 하나 꺼냄

        # 더 내려가면 max_depth를 넘는다면 여기서 멈춤
        if dist[cur] == max_depth:
            continue

        # 현재 노드(cur)의 모든 이웃(자식) 노드를 돈다
        for nb in graph.get(cur, []):
            if nb not in visited:               # 아직 방문 안 했으면
                visited.add(nb)                 # 방문 표시
                parent[nb] = cur                # nb는 cur를 통해 발견되었다고 기록
                dist[nb] = dist[cur] + 1        # 깊이 = 부모 깊이 + 1
                q.append(nb)                    # 나중에 nb도 cur처럼 꺼내서 탐색하도록 큐에 넣기

                # ---- 경로 복원 부분 ----
                # 시작 노드 자체는 빼고(깊이 0) 1단계 이상만 경로로 저장
                if dist[nb] >= 1:
                    path = []                   # 이 nb까지 오는 경로를 담을 리스트
                    x = nb                      # nb에서 시작해서 부모를 따라 위로 올라간다
                    while x is not None:        # 부모가 없을 때(None)까지 반복
                        path.append(x)          # 경로에 하나씩 추가 (뒤에서부터 쌓임)
                        x = parent[x]           # 한 단계 위(부모)로 이동

                    path.reverse()              # 뒤에서부터 쌓였으니 앞에서부터 보이게 뒤집기

                    # 경로 맨 앞에 시작 노드가 있으면 빼준다 (너는 시작 노드 안 보이게 하고 싶다고 했으니까)
                    if path and path[0] == start:
                        path = path[1:]

                    # (이름, 거리, 경로) 형태로 결과에 추가
                    results.append((nb, dist[nb], path))

    return results


def print_results(results, depth, start_name):
    print(f"\n{start_name}으로부터 {depth}단계 이내 감염 경로:")
    for name, d, path in results:
        if path:
            print(f" - {name} (단계 {d}) : {' -> '.join(path)}")
        else:
            print(f" - {name} (단계 {d})")
















# --------- 사용 예시 ----------
infected_person = '채연'  # 시작 노드(감염자)

# 여기는 네가 만든 그래프 이름으로 맞춰 써야 함.
# 예: graph_10node, graph_clean 이런 식으로 쓰면 됨.
node10_res = bfs_with_path(graph_10node, infected_person, max_depth=2)
node100_res = bfs_with_path(graph_100node, infected_person, max_depth=10)

print_results(node10_res, 2, infected_person)
print_results(node100_res, 10, infected_person)
```
