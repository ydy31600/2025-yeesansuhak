```python
from collections import deque

graph = {
    '현빈': ['친구1', '친구2'],
    '친구1': ['친구3'],
    '친구2': ['친구4'],
    '친구3': [],
    '친구4': ['친구1', '친구5'],
    '친구5': ['친구7']
}

def bfs_infection(graph, start, max_depth=2):
    visited = {start}
    dist = {start: 0}
    parent = {start: None}
    q = deque([start])

    results = []

    while q:
        cur = q.popleft()
        if dist[cur] == max_depth:
            continue

        for nb in graph.get(cur, []):
            if nb not in visited:
                visited.add(nb)
                parent[nb] = cur
                dist[nb] = dist[cur] + 1
                q.append(nb)

                if 0 < dist[nb] <= max_depth:
                    path = []
                    x = nb
                    while x is not None:
                        path.append(x)
                        x = parent[x]
                    path.reverse()
                    results.append((nb, dist[nb], path))

    print(f"\n{start}의 감염 경로 (최대 {max_depth}단계, 자기 자신 제외):")
    for target, d, path in results:
        print(f"거리 {d}: " + " -> ".join(path))
#출력
bfs_infection(graph, '현빈', max_depth=2)

```

```python
#결과
from collections import deque

graph = {
    '현빈': ['친구1', '친구2'],
    '친구1': ['친구3'],
    '친구2': ['친구4'],
    '친구3': [],
    '친구4': ['친구1', '친구5'],
    '친구5': ['친구7'],
    '친구7': []
}

def infection_bfs(graph, infected, max_depth=2):  # 2단계까지만 탐색
    queue = deque([(infected, 0)])
    visited = {infected}
    result = []

    while queue:
        person, level = queue.popleft()
        if level == max_depth:
            continue
        for neighbor in graph[person]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append((neighbor, level + 1))
                result.append((neighbor, level + 1))
    return result

infected_person = '현빈'
infected_list = infection_bfs(graph, infected_person)

print(f"감염자: {infected_person}\\n2단계 이내 감염 경로 추적 결과:")
for name, dist in infected_list:
    print(f" - {name} (거리 {dist})")
```
